```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing Raw Data and Header of first 10 apples
data = pd.read_csv("https://raw.githubusercontent.com/cheeseBread54/AP-CSP/main/Apple.csv")
data.head(10)
```

```python
# Tail of Last 10 Apples
data.tail(10)
```

```python
# (Rows, Columns)
data.shape
```

```python
# These columns have floating values
data.describe(include='float64')
```

```python
#These columns have string values
data.describe(include='O')
```

```python
# This trims the data of the sweetness of apple to given the positive values
sweetnessPositiveData = data.query('Sweetness > 0')
```

```python
# Average of the positive data
sweetnessPositiveData.mean()
```

```python
# Apple unit 1: sweetness
display(data.iloc[1,3])
```

```python
# This the whole column of Apple Sweetness
sweetData = data.iloc[0: 4000,3]
display(sweetData)
```

```python
# Most of the apple are average sweetness which is a around 0
n = sweetData
plt.hist(n, bins=100)
plt.title("Apple Sweetness Histogram")
plt.show()

```

```python
# These are histograms are amount(x) vs the different columns(y)
data.hist()
```

```python
#This the whole column of Apple Size
sizeData = data.iloc[0: 4000,1]
display(sizeData)
```

```python
#This the whole column of Apple Weight
weightData = data.iloc[0: 4000,2]
display(weightData)
```

```python
#The correlation of Apple size(x) vs Apple Weight(y)
#This was interesting because it shows the importance of a common value of the apples but it also shows a good demonstration of out lying data points

x1 = sizeData
y1 = weightData
plt.scatter(x1, y1)
plt.title('Apple Size vs Apple Weight Scatter Plot')

plt.axvline(color='black') # y axis zero line colors black
plt.axhline(color='black') # x axis zero line colors black
plt.grid()
plt.xlabel("Apple Size")
plt.ylabel("Apple Weight");

plt.show()

```
